```python
# User-Item data optimizer: reduces the data based on user and item interaction
# thresholds
#   - Total users in the data are reduced by retaining only users with a
#     threshold of interactions. Here the interacion threshold is the users played
#     at least 90 songs.
#   - Total songs in the data are reduced by retaining only songs with a
#     threshold of interactions. Here the interaction threshold is the songs played
#     by at least 120 users

def optimizeDf(option='none'):

  print("REMEDIATION START......")

  if option == 'user':

    print(".Remediating users...")

    if 'df_userOpt' in globals():
       print('.Cleaning up dataframes...')
       global df_userOpt
       del df_userOpt

    # We want our users to have played at least 90 songs to be considered
    USER_SONG_COUNT_CUTOFF = 90
    # Get the column containing the users
    users = df.user_id
    # Create list to hold users to remove based on the song play count cutoff
    remove_users = []
    # Create a dictionary from users to their number of song play count
    play_count = dict()
    dfTotalRows = df.shape[0]
    print("...Total rows in data/dataframe df: ",dfTotalRows)
    print("...Total users before remediation:",len(users))
    print("...Getting list of users and their total song play count...")
    # Create a list that contains each user and their total song play count
    for user in users:
        # If we already have the user, just add 1 to their song play count
        if user in play_count:
            play_count[user] += 1
        # Otherwise, set their song_play_count count to 1
        else:
            play_count[user] = 1
    print("......DONE!")
    print("...Getting list of users with < ",USER_SONG_COUNT_CUTOFF, " song play count...")

    # Create the list of users with less than the song play count cutoff
    for user, num_playCounts in play_count.items():
        if num_playCounts < USER_SONG_COUNT_CUTOFF:
            remove_users.append(user)

    print("......DONE!")
    print("...Total users to remove: ",len(remove_users))
    print("...Removing users with < ",USER_SONG_COUNT_CUTOFF,"song play count...")

    # Remove the users from the df dataframe
    df_userOptimized = df.loc[ ~ df.user_id.isin(remove_users)]

    dfTotalRows =   df_userOptimized.shape[0]
    print("......DONE!")
    print("...Remedation COMPLETE!")
    print("Total rows after remediation: ",dfTotalRows)

    return df_userOptimized

  elif option == 'song':
    print(".Remediating songs...")

    if 'df_final' in globals():
       print('.Cleaning up dataframes...')
       global df_final
       del df_final

    # We want our song to have been played by at least 120 users to be considered
    LISTENER_COUNT_CUTOFF = 120
    # Get the column containing the songs
    songs = df_userOpt.song_id
    # Create list to hold songs to remove based on the listener count cutoff
    remove_songs = []
    # Create a dictionary from songs to their number of listeners
    listener_count = dict()

    dfTotalRows = df_userOpt.shape[0]
    print("...Total rows in dataframe df: ",dfTotalRows)
    print("...Total songs before remediation:",len(songs))
    print("...Getting list of songs and their listener count...")
    # Create the list that contains each song and the number of listeners
    for song in songs:
        # If we already have the song, just add 1 to its listener count
        if song in listener_count:
            listener_count[song] += 1
        # Otherwise, set their listener count to 1
        else:
            listener_count[song] = 1
    print("......DONE!")
    print("...Getting list of songs with < ",LISTENER_COUNT_CUTOFF, " listeners...")

    for song, num_listens in listener_count.items():
        if num_listens < LISTENER_COUNT_CUTOFF:
            remove_songs.append(song)
    print("......DONE!")
    print("...Total songs to remove: ",len(remove_songs))
    print("...Removing songs with < ",LISTENER_COUNT_CUTOFF,"listeners...")

    # Removing songs with less than threshold of listeners and returing the results in a dataframe
    df_songOptimized = df_userOpt.loc[~ df_userOpt.song_id.isin(remove_songs)]

    dfTotalRows =   df_songOptimized.shape[0]
    print("......DONE!")
    print("...Remedation COMPLETE!")
    print(".Total songs after remediation: ",len(df_songOptimized.song_id))
    print(".Total rows after remediation: ",dfTotalRows)
    print("REMEDIATION END")
    return df_songOptimized
  else:
    print("Function optimizeDf : Opimize dataframe for users or songs based on number of plays")
    print("  Usage: optimizeDf('<option>')   where:")
    print("         option : user | song")
    print("  Notes: run optimizeDf('user') first")
```

```python
# Create the bar plot and provide observations
# Chart templates using Python function
# Function to plot a boxplot and a histogram along the same scale

def histogram_boxplot(data, feature, figsize = (10, 7), kde = False, bins = None):

    """
    Boxplot and histogram combined
     data: dataframe
     feature: dataframe column
     figsize: size of figure (default (10, 7))
     kde: whether to the show density curve (default False)
     bins: number of bins for histogram (default None)
    """

    f2, (ax_box2, ax_hist2) = plt.subplots(
        nrows = 2,      # Number of rows of the subplot grid = 2
        sharex = True,  # X-axis will be shared among all subplots
        gridspec_kw = {"height_ratios": (0.25, 0.75)},
        figsize = figsize,
    )  # Creating the 2 subplots
    sns.boxplot(
        data = data, x = feature, ax = ax_box2, showmeans = True, color = "silver", medianprops=dict(color="red", linewidth="3", label='median'), meanprops=dict(label='mean')
    )  # Boxplot will be created and a star will indicate the mean value of the column
    sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2, bins = bins, palette = "pastel"
    ) if bins else sns.histplot(
        data = data, x = feature, kde = kde, ax = ax_hist2
    )  # For histogram
    ax_hist2.axvline(
        data[feature].mean(), color = "green", linestyle = "--"
    )  # Add mean to the histogram
    ax_hist2.axvline(
        data[feature].median(), color = "black", linestyle = "-"
    )  # Add median to the histogram

```

Below is the function to calculate precision@k and recall@k, RMSE, and F1_Score@k to evaluate the model performance.

**Think About It:** Which metric should be used for this problem to compare different models?

```python
def precision_recall_at_k(model, recModel, k=30, threshold=1.5):
    """Return precision and recall at k metrics for each user"""

    # First map the predictions to each user.
    user_est_true = defaultdict(list)

    #Making predictions on the test data
    predictions = model.test(testset)

    for uid, _, true_r, est, _ in predictions:
        user_est_true[uid].append((est, true_r))

    precisions = dict()
    recalls = dict()
    for uid, playing_count in user_est_true.items():

        # Sort play count by estimated value
        playing_count.sort(key=lambda x: x[0], reverse=True)

        # Number of relevant items
        n_rel = sum((true_r >= threshold) for (_, true_r) in playing_count)

        # Number of recommended items in top k
        n_rec_k = sum((est >= threshold) for (est, _) in playing_count[:k])

        # Number of relevant and recommended items in top k
        n_rel_and_rec_k = sum(((true_r >= threshold) and (est >= threshold))
                              for (est, true_r) in playing_count[:k])

        # Precision@K: Proportion of recommended items that are relevant
        # When n_rec_k is 0, Precision is undefined. We here set Precision to 0 when n_rec_k is 0.

        precisions[uid] = n_rel_and_rec_k / n_rec_k if n_rec_k != 0 else 0

        # Recall@K: Proportion of relevant items that are recommended
        # When n_rel is 0, Recall is undefined. We here set Recall to 0 when n_rel is 0.

        recalls[uid] = n_rel_and_rec_k / n_rel if n_rel != 0 else 0

    #Mean of all the predicted precisions are calculated.
    precision = round((sum(prec for prec in precisions.values()) / len(precisions)),3)
    #Mean of all the predicted recalls are calculated.
    recall = round((sum(rec for rec in recalls.values()) / len(recalls)),3)

    # accuracy.rmse(predictions)

    # Update table of accuracy metrics
    recModelRMSE = accuracy.rmse(predictions)
    recModelF1Score = round((2*precision*recall)/(precision+recall),3)
    mask = recModel_df['Model'].eq(recModel)
    recModel_df.loc[mask, ['RMSE', 'Precision','Recall','F1_Score']] = [recModelRMSE, precision,recall,recModelF1Score]

    print('Precision: ', precision) #Command to print the overall precision
    print('Recall: ', recall) #Command to print the overall recall
    print('F_1 score: ', round((2*precision*recall)/(precision+recall),3)) # Formula to compute the F-1 score.
```

Below we will be implementing a function where the input parameters are:

- data: A **song** dataset
- user_id: A user-id **against which we want the recommendations**
- top_n: The **number of songs we want to recommend**
- algo: The algorithm we want to use **for predicting the play_count**
- The output of the function is a **set of top_n items** recommended for the given user_id based on the given algorithm

```python
def get_recommendations(data, user_id, top_n, algo):

    # Creating an empty list to store the recommended song ids
    recommendations = []

    # Creating an user item interactions matrix
    user_item_interactions_matrix = data.pivot(index = 'user_id', columns = 'song_id', values = 'play_count')

    # Extracting those song ids which the user_id has not played yet
    non_interacted_songs = user_item_interactions_matrix.loc[user_id][user_item_interactions_matrix.loc[user_id].isnull()].index.tolist()

    # Looping through each of the song ids which user_id has not interacted yet
    for item_id in non_interacted_songs:

        # Predicting the users for those non played song ids by this user
        est = algo.predict(user_id, item_id).est

        # Appending the predicted play_counts
        recommendations.append((item_id, est))

    # Sorting the predicted play_counts in descending order
    recommendations.sort(key = lambda x: x[1], reverse = True)

    return recommendations[:top_n]  # Returing top n highest predicted play_count songs for this user
```

```python
def ranking_songs(recommendations, playing_count):
  # Sort the songs based on play counts
  ranked_songs = playing_count.loc[[items[0] for items in recommendations]].sort_values('frequency_playCount', ascending = False)[['frequency_playCount']].reset_index()

  # Merge with the recommended songs to get predicted play_counts
  ranked_songs = ranked_songs.merge(pd.DataFrame(recommendations, columns = ['song_id', 'predicted_play_count']), on = 'song_id', how = 'inner')

  # Rank the songs based on corrected play_counts
  ranked_songs['corrected_play_counts'] = ranked_songs['predicted_play_count'] - 1 / np.sqrt(ranked_songs['frequency_playCount'])

  # Sort the songs based on corrected play_counts
  ranked_songs = ranked_songs.sort_values('corrected_play_counts', ascending = False)

  return ranked_songs
```

```python
# Importing necessary packages to work with text data
import nltk
nltk.download('omw-1.4')

# Download punkt library (punctuation)
nltk.download('punkt')

# Download stopwords library
nltk.download('stopwords')

# Download wordnet
nltk.download('wordnet')

# Import regular expression
import re

# Import word_tokenizer
from nltk import word_tokenize

# Import WordNetLemmatizer
from nltk.stem import WordNetLemmatizer

# Import stopwords
from nltk.corpus import stopwords

# Import CountVectorizer and TfidfVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Import cosine_similarity to find the similarity between two vectors
from sklearn.metrics.pairwise import cosine_similarity
```

We will create a **function to pre-process the text data:**

```python
# Create a function to tokenize the text
def tokenize(text):

    # Making each letter as lowercase and removing non-alphabetical text
    text = re.sub(r"[^a-zA-Z]"," ", text.lower())

    # Extracting each word in the text
    tokens = word_tokenize(text)

    # Removing stopwords
    words = [word for word in tokens if word not in stopwords.words("english")]

    # Lemmatize the words
    text_lems = [WordNetLemmatizer().lemmatize(lem).strip() for lem in words]

    return text_lems
```

```python
# Create tfidf vectorizer
tfidf = TfidfVectorizer(tokenizer = tokenize)

# Fit_transfrom the above vectorizer on the text column and then convert the output into an array
song_tfidf = tfidf.fit_transform(df_small['text'].values).toarray()
```

```python
pd.DataFrame(song_tfidf)
```

```python
# Compute the cosine similarity for the tfidf above output
similar_songs = cosine_similarity(song_tfidf, song_tfidf)
similar_songs
```

 Finally, let's create a function to find most similar songs to recommend for a given song.

```python
# Function that takes in song title as input and returns the top 10 recommended songs
def text_recommendations(title, similar_songs):

    recommended_songs = []

    # Getting the index of the song that matches the title
    idx = indices[indices == title].index[0]

    # Creating a Series with the similarity scores in descending order
    score_series = pd.Series(similar_songs[idx]).sort_values(ascending = False)

    # Getting the indexes of the 10 most similar songs
    top_10_indexes = list(score_series.iloc[1 : 11].index)
    print(top_10_indexes)

    # Populating the list with the titles of the best 10 matching songs
    for i in top_10_indexes:
        recommended_songs.append(list(df_small.index)[i])


    return recommended_songs
```
